```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(50, input_dim=11, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(1, activation='linear'))
#model.compile(loss='mse', optimizer='adam', metrics=['mse', 'mae'])

#model.fit(x_train, y_train, epochs=200, validation_split=0.2)
```

```python
import collections
input_spec = collections.OrderedDict([('x', tf.TensorSpec(shape=(None,11), dtype=tf.float64)), ('y', tf.TensorSpec(shape=(None, 1), dtype=tf.float64))])
```

```python
def create_keras_model():
  return Sequential([Dense(50, input_dim=11, activation='relu'), Dense(20, activation='relu'), Dense(1, activation='linear')])
```

```python
def model_fn():
  # We _must_ create a new model here, and _not_ capture it from an external
  # scope. TFF will call this within different graph contexts.
  keras_model = create_keras_model()
  return tff.learning.from_keras_model(keras_model,
      loss=tf.keras.losses.MeanSquaredError(),
      input_spec=input_spec,
      metrics=[tf.keras.metrics.MeanSquaredError()])
```

```python
# Unused
fed_mod = tff.learning.from_keras_model(model,
      loss=tf.keras.losses.MeanSquaredError(),
      input_spec=input_spec,
      metrics=[tf.keras.metrics.MeanSquaredError()])
```

```python
iterative_process = tff.learning.build_federated_averaging_process(
    model_fn,
    client_optimizer_fn=lambda: tf.keras.optimizers.SGD(learning_rate=0.02),
    server_optimizer_fn=lambda: tf.keras.optimizers.SGD(learning_rate=1.0))
```

```python
print(iterative_process.initialize.type_signature.formatted_representation())
```

```python
state = iterative_process.initialize()
```
